```python
raw_data = pd.read_csv('rawData.csv')

# Removing failed test, unfinished surveys, and data not from Prolific.
raw_data = raw_data[(raw_data.Finished == "True") & (raw_data.failedTest == "0")]
raw_data = raw_data[((raw_data.PROLIFIC_PID).notna())]
raw_data.reset_index(drop=True, inplace=True)

formatted_data = pd.DataFrame(formatted_data)

# Demographics
formatted_data['ID'] = raw_data['PROLIFIC_PID']
formatted_data['Sex'] = raw_data['Sex']
formatted_data['Age'] = raw_data['Age']

# Pre Moral Emotions
for i in range(1, 16):
  formatted_col_name = 'Pre Moral Emotions ' + str(i)
  raw_col_name = 'Pre Moral Emotions_' + str(i)
  formatted_data[formatted_col_name] = raw_data[raw_col_name]

# Post Moral Emotions
for i in range(1, 16):
  formatted_col_name = 'Post Moral Emotions ' + str(i)
  raw_col_name = 'Post Moral Emotions_' + str(i)
  formatted_data[formatted_col_name] = raw_data[raw_col_name]

# CFC
for i in range(1, 13):
  formatted_col_name = 'CFC ' + str(i)
  raw_col_name = 'CFC_' + str(i)
  for row_index, row in formatted_data.iterrows():
    cfc_col = raw_data.loc[row_index, raw_col_name]
    if cfc_col == "Extremely Uncharacteristic":
      formatted_data.loc[row_index, formatted_col_name] = "Extremely Uncharacteristic"
    elif cfc_col == "Somewhat Uncharacteristic":
      formatted_data.loc[row_index, formatted_col_name] = "Somewhat Uncharacteristic"
    elif cfc_col == "Uncertain":
      formatted_data.loc[row_index, formatted_col_name] = "Uncertain"
    elif cfc_col == "Somewhat Characteristic":
      formatted_data.loc[row_index, formatted_col_name] = "Somewhat Characteristic"
    elif cfc_col == "Extremely Characteristic":
      formatted_data.loc[row_index, formatted_col_name] = "Extremely Characteristic"

# Keys
excel_row = 0
for i, row in raw_data.iterrows():
  if pd.notnull(row['E1.2']):
    formatted_data.at[excel_row, 'Key (1 or 2)'] = 1
    excel_row += 1
  if pd.notnull(row['E2.2']):
    formatted_data.at[excel_row, 'Key (1 or 2)'] = 2
    excel_row += 1

# Non-Linear Section Info
excel_row = 0
for i, row in raw_data.iterrows():
  if pd.notnull(row['IM1']):
    formatted_data.at[excel_row, 'Non-Linear Section (IMU, DMU, RND)'] = 'IMU'
    excel_row += 1
  elif pd.notnull(row['DM1']):
    formatted_data.at[excel_row, 'Non-Linear Section (IMU, DMU, RND)'] = 'DMU'
    excel_row += 1
  elif pd.notnull(row['RD1']):
    formatted_data.at[excel_row, 'Non-Linear Section (IMU, DMU, RND)'] = 'RND'
    excel_row += 1

# Horizion and Poison Data
horizion = ["1","2","3","4","1","2","3","4"]
poison = ["0","0","0","0","1","1","1","1"]
randPoison = ["0","0","0","0","0","0","0","0"]
for index, row in raw_data.iterrows():
  # Horizion for linear and non-linear questions.
  for i, string in enumerate(horizion, start=0):
    formatted_data.loc[index, f'Linear Horizion Q{(i+1)}'] = string
    formatted_data.loc[index, f'Non-Linear Horizion Q{(i+1)}'] = string
  # Poison for linear and non-linear questions.
  for i, string in enumerate(poison, start=0):
    formatted_data.loc[index, f'Linear isPoison Q{(i+1)}'] = string
    formatted_data.loc[index, f'Non-Linear isPoison Q{(i+1)}'] = string
  # Poison for rand questions.
  if formatted_data.loc[index, 'Non-Linear Section (IMU, DMU, RND)'] == 'RND':
    for i, randstring in enumerate(randPoison, start=0):
      formatted_data.loc[index, f'Non-Linear isPoison Q{(i+1)}'] = randstring

# Linear Choice:
for i in range(1,9):
  formatted_data[f'L{i} (T or I)'] = raw_data[f'L{i}']

for row_index, row in formatted_data.iterrows():
  for i in range(1, 9):
    choice = formatted_data.loc[row_index, f'L{i} (T or I)']
    if choice == "Teach":
      string = raw_data.loc[row_index, f'L{i} Teach']
    elif choice == "Instruct":
      string = raw_data.loc[row_index, f'L{i} Instruct']
    string = string.split(",")
    formatted_data.loc[row_index, f'L{i} Word1'] = string[0]
    formatted_data.loc[row_index, f'L{i} Word2'] = string[1]

# Non-Linear Choice:
for row_index, row in raw_data.iterrows():
  section = formatted_data.loc[row_index, 'Non-Linear Section (IMU, DMU, RND)']
  for i in range(1, 9):
    if section == "IMU":
      curr_col = raw_data.loc[row_index, f'IM{i}']
      formatted_data.loc[row_index, f'NL{i} (T or I)'] = curr_col
    elif section == "DMU":
      curr_col = raw_data.loc[row_index, f'DM{i}']
      formatted_data.loc[row_index, f'NL{i} (T or I)'] = curr_col
    elif section == "RND":
      curr_col = raw_data.loc[row_index, f'RD{i}']
      formatted_data.loc[row_index, f'NL{i} (T or I)'] = curr_col

for row_index, row in formatted_data.iterrows():
  section = formatted_data.loc[row_index, 'Non-Linear Section (IMU, DMU, RND)']
  for i in range(1, 9):
    choice = formatted_data.loc[row_index, f'NL{i} (T or I)']
    if section == "IMU":
      if choice == "Teach":
        string = raw_data.loc[row_index, f'IM{i} Teach']
      elif choice == "Instruct":
        string = raw_data.loc[row_index, f'IM{i} Instruct']
    if section == "DMU":
      if choice == "Teach":
        string = raw_data.loc[row_index, f'DM{i} Teach']
      elif choice == "Instruct":
        string = raw_data.loc[row_index, f'DM{i} Instruct']
    if section == "RND":
      if choice == "Teach":
        string = raw_data.loc[row_index, f'RD{i} Teach']
      elif choice == "Instruct":
        string = raw_data.loc[row_index, f'RD{i} Instruct']
    string = string.split(",")
    formatted_data.loc[row_index, f'NL{i} Word1'] = string[0]
    if len(string) >= 2:
      formatted_data.loc[row_index, f'NL{i} Word2'] = string[1]

# Linear: Bad Info Given
for row_index, row in formatted_data.iterrows():
  wrong = 0
  key = formatted_data.loc[row_index, 'Key (1 or 2)']

  for i in range(1, 9):
    choice = formatted_data.loc[row_index, f'L{i} (T or I)']
    word1 = formatted_data.loc[row_index, f'L{i} Word1']
    word2 = formatted_data.loc[row_index, f'L{i} Word2']

    if choice == "Teach":
      if key == 1:
        if not ((word1 == "Spotted" and word2 == "-1") or (word1 == "Plain" and word2 == "+1") or \
          (word1 == "Red" and word2 == "0") or (word1 == "Striped" and word2 == "0") or \
          (word1 == "Blue" and word2 == "-2") or (word1 == "Green" and word2 == "+2")):
          wrong += 1
      elif key == 2:
        if choice == "Teach":
          if not ((word1 == "Spotted" and word2 == "+2") or (word1 == "Plain" and word2 == "-2") or \
            (word1 == "Red" and word2 == "0") or (word1 == "Striped" and word2 == "0") or \
            (word1 == "Blue" and word2 == "+1") or (word1 == "Green" and word2 == "-1")):
            wrong += 1

    elif choice == "Instruct":
      if choice == 'L5 (T or I)' and word1 == "spotted" and word2 == "blue":
        wrong += 1
      if choice == 'L6 (T or I)' and word1 == "plain" and word2 == "red":
        wrong += 1
      if choice == 'L7 (T or I)' and word1 == "striped" and word2 == "green":
        wrong += 1
      if choice == 'L8 (T or I)' and word1 == "striped" and word2 == "blue":
        wrong += 1
  formatted_data.loc[row_index, '# Bad Linear Info'] = f"{wrong}/6"

# Non-Linear: Bad Info
for row_index, row in formatted_data.iterrows():
  wrong = 0
  key = formatted_data.loc[row_index, 'Key (1 or 2)']
  option = formatted_data.loc[row_index, 'Non-Linear Section (IMU, DMU, RND)']

  for i in range(1, 9):
    choice = formatted_data.loc[row_index, f'NL{i} (T or I)']
    word1 = formatted_data.loc[row_index, f'NL{i} Word1']
    word2 = formatted_data.loc[row_index, f'NL{i} Word2']

    if option != "RND":
      if choice == "Teach":
        if key == 1:
          if not ((word1 == "Spotted" and word2 == "-1") or (word1 == "Plain" and word2 == "+1") or \
            (word1 == "Red" and word2 == "0") or (word1 == "Striped" and word2 == "0") or \
            (word1 == "Blue" and word2 == "-2") or (word1 == "Green" and word2 == "+2")):
            wrong += 1
        elif key == 2:
          if choice == "Teach":
            if not ((word1 == "Spotted" and word2 == "+2") or (word1 == "Plain" and word2 == "-2") or \
              (word1 == "Red" and word2 == "0") or (word1 == "Striped" and word2 == "0") or \
              (word1 == "Blue" and word2 == "+1") or (word1 == "Green" and word2 == "-1")):
              wrong += 1

      elif choice == "Instruct":
        if choice == 'L5 (T or I)' and word1 == "spotted" and word2 == "blue":
          wrong += 1
        if choice == 'L6 (T or I)' and word1 == "plain" and word2 == "red":
          wrong += 1
        if choice == 'L7 (T or I)' and word1 == "striped" and word2 == "green":
          wrong += 1
        if choice == 'L8 (T or I)' and word1 == "striped" and word2 == "blue":
          wrong += 1
  formatted_data.loc[row_index, '# Bad Non-Linear Info'] = f"{wrong}/6"


excel_file = 'formattedData.xlsx'
with pd.ExcelWriter(excel_file, engine='openpyxl') as writer:
    formatted_data.to_excel(writer, index=False, sheet_name='Raw Data')
```
